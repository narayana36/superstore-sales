```python
import pandas as pd
df=pd.read_csv("cleaned_sales_data.csv")
```

```python
df.head()
```

```python
#Total sales
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)
```

```python
#Total Profit
total_profit=df['Profit'].sum()
print("Total Profit:", total_profit)
```

```python
#Sales by region
region_sales=df.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
# Monthly sales Trend 
monthly_sales =df.groupby('Month')['Sales'].sum()
print(monthly_sales)
```

```python
# Every Products total sales
top_products= df.groupby("Category")['Sales'].sum().sort_values(ascending=False).head(5)
print(" Top 5 products: ",top_products)
```

```python
# Top 5 more sales customer names
top_customers=df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(5)
print("Top 5 Customers: ", top_customers)
```
